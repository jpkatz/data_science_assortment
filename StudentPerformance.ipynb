```python
#looking at parents education based on race ethnicity
import pandas as pd
import matplotlib.pyplot as plt
```

```python
students = pd.read_csv('../data/StudentsPerformance.csv',dtype='unicode')
students.head()
```

```python
parents = students[['race/ethnicity','parental level of education']]
parents.head()
```

```python
groupedParentsNormalized=parents.groupby('race/ethnicity')['parental level of education'].value_counts(normalize=True)
```

```python
groupedParentsNormalized.unstack()
```

```python
groupedParentsNormalized.unstack().plot.bar(stacked = True);
```
